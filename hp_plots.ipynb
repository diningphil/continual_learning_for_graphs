```python
import hiplot
import json
```

```python
basedir = "Z:\\CL-DGN\\RESULTS\\GS0\\SuperpixelsBaseline_MNIST_assessment\\HOLDOUT_ASS\\"
```

```python
res = []
for i in range(1, 37):
    with open(basedir + f'HOLDOUT_MS\config_{i}\\config_results.json') as f:
        data = json.load(f)
        curr = {}
        curr['hidden_units'] = data['config']['supervised_config']['hidden_units']
        curr['num_layers'] = data['config']['supervised_config']['num_layers']
        curr['lwf_alpha'] = data['config']['supervised_config']['loss']['args']['alpha']
        curr['lwf_T'] = data['config']['supervised_config']['loss']['args']['distillation_temperature']
        curr['lwf_mode'] = data['config']['supervised_config']['loss']['args']['mode']        
        curr['VL_acc'] = data['VL_main_score']
        res.append(curr)
```

```python
hiplot.Experiment.from_iterable(res).display()
```
